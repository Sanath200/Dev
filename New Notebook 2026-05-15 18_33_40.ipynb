```python
%python
from pyspark.sql import Row

data = [
    Row(EMPID=1, EMPNAME="Alice", MANAGER_ID=None, DEPARTMENT="Executive"),
    Row(EMPID=2, EMPNAME="Bob", MANAGER_ID=1, DEPARTMENT="Finance"),
    Row(EMPID=3, EMPNAME="Charlie", MANAGER_ID=1, DEPARTMENT="HR"),
    Row(EMPID=4, EMPNAME="David", MANAGER_ID=2, DEPARTMENT="Finance"),
    Row(EMPID=5, EMPNAME="Eve", MANAGER_ID=2, DEPARTMENT="Finance"),
    Row(EMPID=6, EMPNAME="Frank", MANAGER_ID=3, DEPARTMENT="HR"),
    Row(EMPID=7, EMPNAME="Grace", MANAGER_ID=3, DEPARTMENT="HR"),
    Row(EMPID=8, EMPNAME="Heidi", MANAGER_ID=4, DEPARTMENT="Finance"),
    Row(EMPID=9, EMPNAME="Ivan", MANAGER_ID=5, DEPARTMENT="Finance"),
    Row(EMPID=10, EMPNAME="Judy", MANAGER_ID=6, DEPARTMENT="HR")
]

emp_hierarchy_df = spark.createDataFrame(data)
display(emp_hierarchy_df)
```

```python
%python
emp_hierarchy_df.createOrReplaceTempView('emp_table')
```

```python
%sql
select * from emp_table
```

```python
%sql
WITH RECURSIVE emp_hierarcy as (
select EMPID,EMPNAME,MANAGER_ID,DEPARTMENT,1 as level,empname as manager from emp_table
where manager_id is null
union all
select e.empid,e.empname,e.MANAGER_ID,e.DEPARTMENT,eh.level+1,concat(eh.manager,',',e.empname) from emp_table e
join emp_hierarcy eh on e.manager_id=eh.empid
)
select * from emp_hierarcy
```
