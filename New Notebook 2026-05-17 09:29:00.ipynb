```python
from pyspark.sql import Row

employee_data = [
    Row(employee_id=1, name="Alice", department="HR", salary=70000, hire_date="2020-01-15"),
    Row(employee_id=2, name="Bob", department="Engineering", salary=95000, hire_date="2019-03-22"),
    Row(employee_id=3, name="Charlie", department="Finance", salary=80000, hire_date="2021-07-01"),
    Row(employee_id=4, name="Diana", department="Engineering", salary=99000, hire_date="2018-11-12"),
    Row(employee_id=5, name="Eve", department="HR", salary=72000, hire_date="2022-05-30")
]

df_employee = spark.createDataFrame(employee_data)
display(df_employee)
```

```python
df_employee.show()
```

```python
from pyspark.sql.functions import col,sum
df_employee.groupBy('department').agg(sum(col("salary")).alias("total_salary")).display()
```

```python
df_employee.explain()
```

```python
dept_data = [
    Row(dept_id=1, dept_name="HR"),
    Row(dept_id=2, dept_name="Engineering"),
    Row(dept_id=3, dept_name="Finance")
]

df_dept = spark.createDataFrame(dept_data)
display(df_dept)
```

```python
join = df_employee.join(df_dept.hint('merge'), df_employee.department == df_dept.dept_name, 'inner')
```

```python
spark.conf.get("spark.sql.shuffle.partitions")
```

```python
join.display()
```

```python
spark.conf.set('spark.sql.files.maxPartitionBytes', '512mb')
```

```python
config_value = spark.conf.get('spark.sql.files.maxPartitionBytes')
value_bytes = int(config_value.rstrip('b'))
value_mb = value_bytes / (1024 * 1024)
print(f"{value_bytes} bytes means {value_mb:.0f} MB")
```
