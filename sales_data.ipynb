```python
# 1. Sample sales data
data = [
    (1, "2026-05-01", "Laptop", 2, 80000),
    (2, "2026-05-01", "Mobile", 5, 20000),
    (3, "2026-05-02", "Tablet", 3, 15000),
    (4, "2026-05-02", "Laptop", 1, 80000),
    (5, "2026-05-03", "Headphones", 4, 2000),
    (6, "2026-05-03", "Mobile", 2, 22000),
    (7, "2026-05-04", "Tablet", 1, 15000),
    (8, "2026-05-04", "Laptop", 3, 78000),
    (9, "2026-05-05", "Headphones", 6, 2500),
    (10, "2026-05-05", "Mobile", 2, 21000)
]

# 2. Create DataFrame
columns = ["sale_id", "sale_date", "product", "quantity", "price"]
df = spark.createDataFrame(data, columns)

# 3. Add total column
from pyspark.sql.functions import col
df = df.withColumn("total_amount", col("quantity") * col("price"))


# 5. Save as table (Delta)
df.write.mode("overwrite").saveAsTable("first_catalog.default.sales_data")

# 6. Read back
df_read = spark.read.table("first_catalog.default.sales_data")
df_read.display()
```
